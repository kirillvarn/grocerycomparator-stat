```python
top_ridge[['dependent', 'independent_parameters', 'score']].to_excel('ridgetop.xlsx')
top_lasso[['dependent', 'independent_parameters', 'score']].to_excel('lassotop.xlsx')
top_regr[['dependent', 'independent_parameters', 'score']].to_excel('lineartop.xlsx')
```

```python
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import LassoCV
```

```python
names = ['lasso', 'ridge', 'linear']
for dataset_id, dataset in enumerate([top_lasso, top_ridge, top_regr]):
    fig, ax = plt.subplots(1, 3)
    fig.set_size_inches(11.1, 4)
    for idx, item in enumerate(dataset.iterrows()):
        row = item[1]
        # print(item[1]['dependent'])
        x = prices[row['independent_parameters']]
        y = prices[row['dependent']]

        x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=0, shuffle=False)
        ml_model = LinearRegression()
        if names[dataset_id] == "linear":
            ml_model = LinearRegression()

        if names[dataset_id] == "ridge":
            ml_model = RidgeCV()

        if names[dataset_id] == "lasso":
            ml_model = LassoCV()

        ml_model.fit(x_train, y_train)
        y_pred = ml_model.predict(x_test)
        score = r2_score(y_test, y_pred)
        print(f"{row['dependent']}: {score}")
        ax[idx].plot(range(len(y_test)), y_test)
        ax[idx].plot(range(len(y_pred)), y_pred)
        ax[idx].legend(['Actual', 'Predicted'])
        ax[idx].set_xlabel(row['dependent'][8:])
    ax[0].set_ylabel("Price")
        # plt.ylabel("Pitsa salaami ja chorizo Feliciana 320g")
        # plt.xlabel("Time")
    # plt.tick_params(labelbottom=False)
    plt.tight_layout()

    plt.savefig(f'{names[dataset_id]}.png')
    # plt.close()
    # score = r2_score(y_test, y_pred)
    # ax.lineplot()
    # i.plot()
```
